**PROBLEM: MAKİNE SATIN ALMA VE OPTİMİZASYON PROBLEMİ**

Bir şirket, yeni ürün üretmek için makine satın almak istemektedir. Şirketin makine için ayırdığı para $800.000\text{ TL}$'dir. Makine imalatçısı firma, şirkete üç tip model önerir. Klasik modelin fiyatı $24.000\text{ TL}$, lüks modelin fiyatı $40.000\text{ TL}$ ve süper modelin ise fiyatı $80.000\text{ TL}$'dir. Klasik model $15\text{ m}^2$, lüks model $20\text{ m}^2$, süper model $30\text{ m}^2$ giriş katında yer tutar. Makinelerin üretimi ve onları çalıştırmak için gerekli olan operatör sayıları aşağıdaki gibidir:

| Model Tipi | Üretim Miktarı / Saat | Operatör Sayısı |
| :--- | :---: | :---: |
| Klasik model | $4$ | $1$ |
| Lüks model | $5$ | $2$ |
| Süper model | $20$ | $5$ |

Şirket satın alınacak makinelerin en az yarısının lüks model olması üzerinde durmaktadır. Makinelerin ürettiği ürün başına operatörün maliyeti dışında $80\text{ TL}$'lik brüt bir kâr beklenmektedir. Üretilen tüm ürünler birim başına $80\text{ TL}$'lik bir kârla satılabilecektir. Şirket giriş katında makineler için $500\text{ m}^2$ yer ve onları çalıştıracak $100$ operatör ayırabiliyor. Operatörlere saatte $30\text{ TL}$ ödenmektedir. Şirketin amacı kârını ençoklaması için klasik, lüks ve süper makinelerden satın alacağı sayıları belirlemektir. Problemi buna göre doğrusal programlama modeli olarak formülleştiriniz.

**Problemin Matematiksel Modeli**

***1. Karar Değişkenleri***

$x_1$: Satın alınacak klasik makine sayısı<br>$x_2$: Satın alınacak lüks makine sayısı<br>$x_3$: Satın alınacak süper makine sayısı

***2. Amaç Fonksiyonu (Saatlik Net Kârı Enbüyükleme)***

Her bir makinenin saatlik net kârı: $\text{Net K\^ar} = (\text{\"Uretim/Saat} \times 80\text{ TL}) - (\text{Operat\"or Sayisi} \times 30\text{ TL})$ formülüyle hesaplanır.<br><br>Klasik ($x_1$): $(4 \times 80) - (1 \times 30) = 290\text{ TL/saat}$<br>Lüks ($x_2$): $(5 \times 80) - (2 \times 30) = 340\text{ TL/saat}$<br>Süper ($x_3$): $(20 \times 80) - (5 \times 30) = 1450\text{ TL/saat}$$$\max z = 290x_1 + 340x_2 + 1450x_3$$

***3. Kısıtlayıcılar***

Bütçe Kısıtı:$$24.000x_1 + 40.000x_2 + 80.000x_3 \le 800.000$$Yer (Alan) Kısıtı:$$15x_1 + 20x_2 + 30x_3 \le 500$$İş Gücü (Operatör) Kısıtı:$$1x_1 + 2x_2 + 5x_3 \le 100$$Lüks Model Oranı Kısıtı (En az $\%50$):$$\frac{x_2}{x_1 + x_2 + x_3} \ge 0.5 \implies -0.5x_1 + 0.5x_2 - 0.5x_3 \ge 0$$Doğal Kısıtlar (Negatif Olmama ve Tam Sayı):$$x_1, x_2, x_3 \ge 0 \quad \text{ve hepsi tam sayı olmalıdır.}$$

**Modelin Optimal Çözümü ve Analizi**

***Senaryo A: Sürekli Doğrusal Programlama Çözümü (Ondalıklı)***

Değişkenlerin tam sayı olma zorunluluğu dikkate alınmadan (makinelerin bölünebilir olduğu varsayılarak) yapılan teorik çözüm:<br>$x_1 = 0$<br>$x_2 = 6.67$<br>$x_3 = 6.67$<br>$\max z = 11.933,33\text{ TL/saat}$<br>Değerlendirme: Matematiksel olarak kârı en tepeye çıkaran nokta burası olsa da gerçek hayatta bir şirket mağazadan $6.67$ adet makine satın alamaz. Dolayısıyla bu çözüm pratik açıdan geçersizdir.

***Senaryo B: Tam Sayılı Doğrusal Programlama Çözümü (Gerçekçi)***

Makinelerin tam adetlerde ($1, 2, 3...$) alınması kısıtı altında (Gomory Kesme veya Dal-Sınır algoritmaları kullanılarak) yapılan nihai çözüm:<br>$x_1 = 0$ (Klasik makineden hiç alınmaz)<br>$x_2 = 8$ (8 Adet Lüks makine alınır)<br>$x_3 = 6$ (6 Adet Süper makine alınır)<br>$\max z = 11.420\text{ TL/saat}$

***Nihai Çözümün Kontrolü (Kısıt Doğrulaması):***

*Bütçe Kontrolü:* <br>$(8 \times 40.000) + (6 \times 80.000) = 320.000 + 480.000 = 800.000\text{ TL} \le 800.000\text{ TL}$ (Tam bütçe sınırında, kısıt sağlandı).<br><br>*Alan Kontrolü:* <br>$(8 \times 20) + (6 \times 30) = 160 + 180 = 340\text{ m}^2 \le 500\text{ m}^2$ (Alan kısıtı sağlandı, $160\text{ m}^2$ boş alan kaldı).<br><br>*Operatör Kontrolü:* <br>$(8 \times 2) + (6 \times 5) = 16 + 30 = 46 \text{ kişi} \le 100 \text{ kişi}$ (Operatör kısıtı sağlandı).<br><br>*Lüks Oranı Kontrolü:* <br>Toplam $8 + 6 = 14$ makine vardır. <br><br>*Lüks makine oranı:* <br>$8 / 14 = \%57.1$'dir. Bu oran, hedeflenen $\%50$ sınırının üzerinde olduğu için kısıt başarıyla sağlanmıştır.

Sonuç olarak; Şirket saatlik net kârını enbüyüklemek için 8 adet lüks, 6 adet süper makine satın almalı ve klasik makineden hiç almamalıdır. Bu yatırım sonucunda şirket saatte $11.420\text{ TL}$ net kâr elde edecektir.

```python
import os
import sys
import pandas as pd
from scipy.optimize import linprog

def makine_optimizasyonu_coz(excel_yolu="makine_verileri.xlsx"):
    print("=" * 70)
    print("DİNAMİK MAKİNE OPTİMİZASYONU ÇÖZÜCÜ")
    print("=" * 70)

    # 1. Excel Dosyasının Klasörde Olup Olmadığının Kontrolü
    if not os.path.exists(excel_yolu):
        print(f"\n[!] Hata: '{excel_yolu}' dosyası bulunamadı!")
        print("Lütfen oluşturduğunuz Excel dosyasının bu Python koduyla aynı")
        print("klasörde yer aldığından emin olun.")
        print("=" * 70)
        return

    # 2. Excel Sayfalarını Okuma
    try:
        df_makine = pd.read_excel(excel_yolu, sheet_name="Makine_Ozellikleri")
        df_param = pd.read_excel(excel_yolu, sheet_name="Kisitlar_ve_Parametreler")
    except Exception as e:
        print(f"\n[!] Excel okunurken bir hata oluştu: {e}")
        print("Lütfen Excel'deki sayfa isimlerinin 'Makine_Ozellikleri' ve")
        print("'Kisitlar_ve_Parametreler' olduğundan emin olun.")
        print("=" * 70)
        return

    # 3. Parametreleri Kolay Erişim İçin Sözlük (Dictionary) Yapısına Dönüştürme
    try:
        # Sütun isimlerinin sağındaki/solundaki boşlukları temizleyerek sözlüğe aktarıyoruz
        parametreler = dict(zip(df_param["Parametre"].str.strip(), df_param["Deger"]))
    except Exception as e:
        print(f"\n[!] Parametreler okunurken hata oluştu: {e}")
        print("Lütfen parametre sayfanızda 'Parametre' ve 'Deger' sütunlarının olduğunu kontrol edin.")
        return

    # 4. Gerekli Parametrelerin Kontrolü
    gerekli_param_listesi = [
        "Toplam_Butce_TL", "Toplam_Alan_m2", "Maks_Operator_Sayisi",
        "Minimum_Luks_Orani", "Urun_Birim_Kar_TL", "Operator_Saatlik_Ucret_TL"
    ]
    
    for param in gerekli_param_listesi:
        if param not in parametreler:
            print(f"\n[!] Hata: Kısıtlar sayfasında '{param}' parametresi eksik veya hatalı yazılmış!")
            return

    # 5. Dinamik Net Kâr Hesaplama (Amaç Fonksiyonu Katsayıları)
    birim_kar = parametreler["Urun_Birim_Kar_TL"]
    op_ucret = parametreler["Operator_Saatlik_Ucret_TL"]

    print("\nHesaplanan Saatlik Net Kâr Değerleri:")
    c = []
    for idx, row in df_makine.iterrows():
        # Saatlik Net Kâr = (Üretim Miktarı * Birim Kâr) - (Operatör Sayısı * Saatlik Ücret)
        net_kar = (row["Uretim_Miktari_Saat"] * birim_kar) - (row["Operator_Sayisi"] * op_ucret)
        c.append(-net_kar)  # scipy.optimize.linprog varsayılan olarak minimize ettiği için eksi ile çarpıyoruz
        print(f"  - {row['Makine_Tipi']} Makine Net Kârı: {net_kar:.2f} TL/saat")

    # 6. Kısıt Matrislerinin İnşası (A_ub ve b_ub)
    A_ub = []
    b_ub = []

    # A) Bütçe Kısıtı (<= Toplam_Butce_TL)
    A_ub.append(df_makine["Fiyat_TL"].tolist())
    b_ub.append(parametreler["Toplam_Butce_TL"])

    # B) Alan Kısıtı (<= Toplam_Alan_m2)
    A_ub.append(df_makine["Alan_m2"].tolist())
    b_ub.append(parametreler["Toplam_Alan_m2"])

    # C) Operatör Kısıtı (<= Maks_Operator_Sayisi)
    A_ub.append(df_makine["Operator_Sayisi"].tolist())
    b_ub.append(parametreler["Maks_Operator_Sayisi"])

    # D) Lüks Oranı Kısıtı (Lüks Sayısı >= Minimum_Luks_Orani * Toplam Makine Sayısı)
    # Formülün Standart Formata Dönüştürülmesi: r*x1 + (r - 1)*x_luks + r*x3 + ... <= 0
    r = parametreler["Minimum_Luks_Orani"]
    luks_kisit_satiri = []
    for idx, row in df_makine.iterrows():
        makine_adi_temiz = str(row["Makine_Tipi"]).strip().lower()
        if makine_adi_temiz == "lüks" or makine_adi_temiz == "luks":
            luks_kisit_satiri.append(r - 1.0)
        else:
            luks_kisit_satiri.append(r)
    
    A_ub.append(luks_kisit_satiri)
    b_ub.append(0.0)

    # Değişken sınırları (Her makine için x >= 0)
    bounds = [(0, None) for _ in range(len(df_makine))]

    # -----------------------------------------------------------------
    # A) SÜREKLİ DOĞRUSAL PROGRAMLAMA ÇÖZÜMÜ (Continuous LP)
    # -----------------------------------------------------------------
    res_lp = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method='highs')
    
    print("\n" + "-" * 50)
    print("[A] SÜREKLİ DOĞRUSAL PROGRAMLAMA (LP) ÇÖZÜMÜ")
    print("-" * 50)
    if res_lp.success:
        for idx, row in df_makine.iterrows():
            print(f"  - Satın Alınacak {row['Makine_Tipi']} Makine Adedi: {res_lp.x[idx]:.4f}")
        print(f"  - Toplam Saatlik Net Kâr: {-res_lp.fun:.2f} TL")
    else:
        print("  - Sürekli doğrusal programlama modeli için uygun çözüm bulunamadı!")

    # -----------------------------------------------------------------
    # B) TAM SAYILI DOĞRUSAL PROGRAMLAMA ÇÖZÜMÜ (MIP / ILP)
    # -----------------------------------------------------------------
    # Tam sayılı çözüm elde etmek için tüm makine değişkenlerine tam sayı (integrality=1) kısıtı veriyoruz.
    integrality = [1] * len(df_makine)
    res_ilp = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, integrality=integrality, method='highs')

    print("\n" + "-" * 50)
    print("[B] TAM SAYILI DOĞRUSAL PROGRAMLAMA (ILP) ÇÖZÜMÜ")
    print("-" * 50)
    if res_ilp.success:
        for idx, row in df_makine.iterrows():
            # Tam sayı çözümü olduğundan değeri en yakın tam sayıya yuvarlayarak int formatında yazdırıyoruz
            adet = int(round(res_ilp.x[idx]))
            print(f"  - Satın Alınacak {row['Makine_Tipi']} Makine Adedi: {adet}")
        print(f"  - Toplam Saatlik Net Kâr: {-res_ilp.fun:.2f} TL")
    else:
        print("  - Tam sayılı model için uygun bir çözüm bulunamadı!")
    print("=" * 70)


if __name__ == "__main__":
    # Gerekli harici kütüphanelerin kontrolü
    try:
        import pandas as pd
        import openpyxl
        import scipy
    except ImportError:
        print("[!] Gerekli kütüphaneler bilgisayarınızda kurulu değil.")
        print("Lütfen terminalde aşağıdaki komutu çalıştırın:")
        print("pip install pandas openpyxl scipy numpy")
        sys.exit()

    # Optimizasyon fonksiyonunu çalıştır
    makine_optimizasyonu_coz("makine_verileri.xlsx")
```

<br>Kitaptaki teorik çözüm adımları ile Python (scipy.optimize.linprog) kütüphanesi kullanılarak elde edilen dijital çözüm çıktıları birebir örtüşmekte ve doğrusal modelin başarısını kanıtlamaktadır. Sürekli Doğrusal Programlama (LP) modelinde, her iki yöntem de klasik makineden hiç alınmaması gerektiğini ($x_1 = 0$), lüks ve süper makinelerden ise kesirli olarak $6.67$ adet alınarak saatlik maksimum $11.933,33\text{ TL}$ kâr elde edileceğini hesaplamıştır. Gerçek dünya şartlarına uygun olan Tam Sayılı Doğrusal Programlama (ILP) modeline geçildiğinde ise; hem kitaptaki manuel kısıt kontrol adımları hem de Python'ın integrality parametresiyle çalışan gelişmiş Branch-and-Bound (Dal-Sınır) algoritması, şirketi en optimum kombinasyon olan 8 adet Lüks ($x_2$) ve 6 adet Süper ($x_3$) makine alımına yönlendirmiştir. Bu gerçekçi senaryoda, toplam saatlik net kârın $11.420\text{ TL}$ olacağı ve şirketin politikası olan "en az %50 lüks makine barındırma" kısıtının da (satın alınan makinelerin %57'si lüks olacak şekilde) iki çözümde de güvenli bir şekilde sağlandığı doğrulanmıştır.
